```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown
from gtts import gTTS
from IPython.display import Audio,display


def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

```

```python
def ask_question(name):
  ques = ' hey '  +  name  + ' what do you want  '
  ques = input(ques)
  return ques
```

```python
def classify_questions(ques):
    goahead_with_web_browsing = False
    device_lst = ['email', 'message', 'call', 'reminder', 'calculator', 'alarm']
    personal_lst = ['who are you?', 'who created you']

    device = False
    for i in device_lst:
        if i in ques.lower():
            device = True

    if device:
        print('This question is related to a device, which is not supported in this Intellivoice')
        goahead_with_web_browsing = False

    personal_question = False
    for i in personal_lst:
        if i in ques.lower():
            personal_question = True

    if personal_question:
        print('Hey, I am Intellivoice created by Sai')

    if device or personal_question:
        goahead_with_web_browsing = False
    else:
        goahead_with_web_browsing = True

    return goahead_with_web_browsing

classify_questions('who are you?')
```

```python
def ask_promptpal(ques):
  modified_prompt = '  hey give me answer to this question '+  ques  + ' in a maximum of 80 words  '
  response = model.generate_content(modified_prompt)
  to_markdown(response.text)
  return response.text
```

```python
def speak(answer):

  tts = gTTS(answer)
  tts.save('audio.mp3')
  display(Audio('audio.mp3',autoplay=True))
```

```python
have_any_other_queries = 'y'
name = ''
while have_any_other_queries=='y':
  if name =='':
    name=input('hey what is your name ? -')

  q = ask_question(name)
  go_ahead = classify_questions(q)
  answer = ''


  if go_ahead == True:
    answer = ask_promptpal(q)
    speak(answer)

    print(answer)

  have_any_other_queries = input(' Do you have any other queries ')
```
